<h2>Import Libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

<h2>Import Data

```python
df = pd.read_excel("Olympics_100m_History.xlsx")
df
```

```python
#Sort data by year
df = df.sort_values("Year")
df
```

```python
#Subset data, removing the Winner column
df = df[['Year', 'Time', 'Gender']]
df
```

```python
#Check data types
df.dtypes
```

```python
#Create line graph that shows the historical trend of the Olympic 100m
ax = sns.lineplot(x = "Year", y = "Time", data=df, hue="Gender")
ax.set_title('1896-2021 Olympic 100m Champ Progression')
plt.show()
```

```python
#Create data frame filtered on women
df_women = df.query('Gender == "Women"')
df_women
```

```python
#Create dataframe filtered on men
df_men = df.query('Gender == "Men"')
df_men
```

```python
#Subset df_women
df_women = df_women[['Year','Time']]
df_women
```

```python
#Subset df_men
df_men = df_men[['Year','Time']]
df_men
```

```python
#Set Year as index
df_women = df_women.set_index("Year")
df_women
```

```python
#Set Year as index
df_men = df_men.set_index("Year")
df_men
```

<h2>Import metrics and models, set up training/test split

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error
from statsmodels.tsa.statespace.sarimax import SARIMAX
from statsmodels.tsa.arima.model import ARIMA
```

```python
train_women = df_women.iloc[:15]
test_women = df_women.iloc[-6:]
```

```python
train_men = df_men.iloc[:24]
test_men = df_men.iloc[-6:]
```

```python
train_women
```

```python
train_men
```

```python
test_women
```

```python
test_men
```

```python
arima_model_women = ARIMA(train_women, order=(1,1,0))
arima_result_women = arima_model_women.fit()
```

```python
arima_model_men = ARIMA(train_men, order=(1,1,0))
arima_result_men = arima_model_men.fit()
```

```python
arima_forecast_women = arima_result_women.forecast(steps=6)
arima_forecast_women
```

```python
arima_forecast_men = arima_result_men.forecast(steps=6)
arima_forecast_men
```

```python
sarimax_model_women = SARIMAX(train_women, order=(2,1,0))
sarimax_result_women = sarimax_model_women.fit()
sarimax_forecast_women = sarimax_result_women.forecast(steps=1)
sarimax_forecast_women
```

```python
sarimax_model_men = SARIMAX(train_men, order=(2,1,0))
sarimax_result_men = sarimax_model_men.fit()
sarimax_forecast_men = sarimax_result_men.forecast(steps=1)
sarimax_forecast_men
```

<h2>Women's 100M Winning Time Prediction

```python
arima_model_org_women = ARIMA(df_women, order=(2,1,0))
arima_result_women = arima_model_org_women.fit()
arima_forecast_women = arima_result_women.forecast(steps=1)
arima_forecast_women
```

<h2>Men's 100M Winning Time Prediction

```python
arima_model_org_men = ARIMA(df_men, order=(2,1,0))
arima_result_men = arima_model_org_men.fit()
arima_forecast_men = arima_result_men.forecast(steps=1)
arima_forecast_men
```

```python
arima_forecast_women_df = pd.DataFrame(arima_forecast_women)
arima_forecast_women_df
```

```python
arima_forecast_men_df = pd.DataFrame(arima_forecast_men)
arima_forecast_men_df
```

```python
arima_forecast_women_df['Year'] = '2024(p)'
arima_forecast_women_df['Gender'] = 'Women'
arima_forecast_women_df
```

```python
arima_forecast_men_df['Year'] = '2024(p)'
arima_forecast_men_df['Gender'] = 'Men'
arima_forecast_men_df
```

```python
arima_forecast_women_df = arima_forecast_women_df.rename(columns={0:'Time'})
arima_forecast_women_df
```

```python
arima_forecast_men_df = arima_forecast_men_df.rename(columns={0:'Time'})
arima_forecast_men_df
```

```python
arima_forecast_women_df['Time'] = round(arima_forecast_women_df['Time'],2)
arima_forecast_women_df = arima_forecast_women_df[['Year','Time','Gender']]
arima_forecast_women_df
```

```python
arima_forecast_men_df['Time'] = round(arima_forecast_men_df['Time'],2)
arima_forecast_men_df = arima_forecast_men_df[['Year','Time','Gender']]
arima_forecast_men_df
```

```python
final_df = pd.concat([df, arima_forecast_women_df, arima_forecast_men_df])
final_df
```

```python
final_df = final_df.reset_index(drop=True)
final_df
```
